Thêm các thư viện cần thiết ở đây

```python
# Import các thư viện cần thiết cho bài toán

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
```

Load dữ liệu










```python
# Đọc dữ liệu từ sklearn

cancer_set = load_breast_cancer()

# Chuyển thành DataFrame

cancer_data = pd.DataFrame(data=cancer_set ['data'], columns=cancer_set['feature_names'])

y = cancer_set.target
```

```python
# Do range của mỗi cột khá lệch nhau nên ta sử dụng chuẩn hóa dữ liệu

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

Chuẩn hóa dữ liệu đầu vào


```python
# Fit vào dữ liệu
scaler.fit(cancer_data)

# Thực hiện transform scale
scale_cancer_data = scaler.transform(cancer_data)
```

```python
X_train, X_test, y_train, y_test =train_test_split(scale_cancer_data, y, test_size = 0.2, random_state = 30)

from sklearn.linear_model import LogisticRegression

Model  = LogisticRegression()

Model.fit(X_train, y_train)

Model.score(X_test, y_test)
```

Thực hiện PCA

```python
from sklearn.decomposition import PCA

# Khởi tạo đối tượng PCA với số comp = 2
my_pca = PCA (n_components = 2 )

# Fit vào data
my_pca.fit(scale_cancer_data)

# Thực hiện transform 
pca_scale_cancer_data = my_pca.transform(scale_cancer_data)
```

```python
X_train, X_test, y_train, y_test =train_test_split(pca_scale_cancer_data, y, test_size = 0.2, random_state = 30)

from sklearn.linear_model import LogisticRegression

Model  = LogisticRegression()

Model.fit(X_train, y_train)

Model.score(X_test, y_test)
```

```python
plt.figure(figsize = (10,6))
# Thành phần comp số 1
pca_1 = pca_scale_cancer_data[:, 0]
# Thành phần comp số 2
pca_2 = pca_scale_cancer_data[:, 1]
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Scatter Plot of First Two Principal Components')
# Vẽ đồ thị
plt.scatter(x=pca_1, y = pca_2, c= cancer_set['target'])
```

```python
fig = plt.figure(1, figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d", elev=-150, azim=110)

X_reduced = PCA(n_components=3).fit_transform(scale_cancer_data.data)

ax.scatter(
    X_reduced[:, 0],
    X_reduced[:, 1],
    X_reduced[:, 2],
    c= cancer_set['target'],
    cmap=plt.cm.Set1,
    edgecolor="k",
    s=40,
)

ax.set_title("First three PCA directions")
ax.set_xlabel("1st eigenvector")
ax.xaxis.set_ticklabels([])
ax.set_ylabel("2nd eigenvector")
ax.yaxis.set_ticklabels([])
ax.set_zlabel("3rd eigenvector")
ax.zaxis.set_ticklabels([])

plt.show()
```
